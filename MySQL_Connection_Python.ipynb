```python
import csv
```

```python
with open('glass.data','r') as f:                 # this line open file which is uploaded jupyternote book
    glass_data = csv.reader(f, delimiter='\n')    # this line read out file line by line bcz we defined delimiter = '\n'
    print(enumerate(glass_data))        # enumarate always gives indexing to each and every record this line just for understand, generally enumarate means number of things one bye one 
    for lines in enumerate(glass_data): # this line fetech data index-wise
        for list_ in (lines[1]):        # this loop insert values of list_ into a table glassdata using curseor.execute(..) code
            cursor.execute('insert into glassdata values ({values})'.format(values=list_))
    mydb.commit() # it used to commit that these recoreds have to inserted in table
```
