### Feature Selection
How do I decide **which features have to be included** in a linear model? Here's one idea:
- Try different models, and only keep predictors in the model if they have small p-values.
- Check if the R-squared value goes up when you add new predictors to the model.

What are the **drawbacks** in this approach?
-If the underlying assumptions for creating a Linear model(the features being independent) are violated(which usually is the case),p-values and R-squared values are less reliable.
- Using a p-value cutoff of 0.05 means that adding 100 predictors to a model that are **pure noise**, still 5 of them (on average) will be counted as significant.
- R-squared is susceptible to **model overfitting**, and thus there is no guarantee that a model with a high R-squared value will generalise. Following is an example:

```python
lm = smf.ols(formula='Price ~  SqFt', data=data).fit()
lm.rsquared
```

```python
lm = smf.ols(formula='Price ~  SqFt+Bedrooms', data=data).fit()
print("rsquared",lm.rsquared)
print("adusted-rsquared",lm.rsquared_adj)
```

```python
lm = smf.ols(formula='Price ~  SqFt+Bedrooms+Bathrooms', data=data).fit()
print("rsquared",lm.rsquared)
print("adusted-rsquared",lm.rsquared_adj)
```

```python
lm = smf.ols(formula='Price ~  SqFt+Bedrooms+Bathrooms+Offers', data=data).fit()
print("rsquared",lm.rsquared)
print("adusted-rsquared",lm.rsquared_adj)
```

Selecting the model with the highest value of R-squared is not a correct approach as the value of R-squared shall always increase whenever a new feature is taken for consideration even if the feature is unrelated to the response.

The alternative is to use **adjusted R-squared** which penalises the model complexity (to control overfitting), but this again generally under-penalizes complexity.

a better approach to feature selection is **Cross-validation**. It provides a more reliable way to choose which of the created models will best **generalise** as it better estimates of out-of-sample error. An advantage is that the cross-validation method can be applied to any machine learning model and the scikit-learn package provides extensive functionality for that.

### Handling Categorical variables with More than Two Categories

Let's create a new column called **Neighborbood_Direction**, and randomly assign observations to be **Eest, West, South and North**:
### We need to represent the ‘Neighborbood_Direction’ column numerically. But mapping East=0, West=1, North=2 and South = 3 will mean that rural is three times south which is not the case. Hence, we’ll create another **dummy variable**:
# One Hot encoding

```python

# set a seed for reproducibility
np.random.seed(123456)


# assign roughly one third of observations to each group
nums = np.random.rand(len(data))
mask_West = (nums>0.25) & (nums<0.50)
mask_South = (nums>0.50) & (nums<0.75)
mask_North = nums>0.75
data["Neighborhood_Direction"] = "East"
data.loc[mask_West, "Neighborhood_Direction"] = "West"
data.loc[mask_South,"Neighborhood_Direction"] = "South"
data.loc[mask_North,'Neighborhood_Direction']='North'
data.head()
```

```python
# create three dummy variables using get_dummies, then exclude the first dummy column
area_dummies = pd.get_dummies(data["Neighborhood_Direction"], prefix='Neighborhood_Direction').iloc[:,1:]

# concatenate the dummy variable columns onto the original DataFrame (axis=0 means rows, axis=1 means columns)
data = pd.concat([data, area_dummies], axis=1)
data.head()
```

What does the encoding say?

East is encoded as Neighborhood_Direction_North	=0 , Neighborhood_Direction_South=0 and Neighborhood_Direction_West= 0
West is encoded as Neighborhood_Direction_North	=0 , Neighborhood_Direction_South=0 and Neighborhood_Direction_West= 1
North is encoded as Neighborhood_Direction_North=1 , Neighborhood_Direction_South=0 and Neighborhood_Direction_West= 0
South is encoded as Neighborhood_Direction_North=0 , Neighborhood_Direction_South=1 and Neighborhood_Direction_West= 0

Now the question is: **Why have we used 3 dummy columns instead of 4?**

Because using only three dummy columns, we can capture the information of all the 4 columns. For example, if the value for  Neighborhood_Direction_North, Neighborhood_Direction_South as well as Neighborhood_Direction_West is 0, it automatically means that the data belongs toNeighborhood_Direction_East.

This is called handling the dummy variable trap. If there are N dummy variable columns, then the same information can be conveyed by N-1 columns. Let's include the three new dummy variables in the model:

```python
features_cols=["SqFt","Bedrooms","Bathrooms","Offers","IsBrick","Neighborhood_Direction_North","Neighborhood_Direction_South","Neighborhood_Direction_West"] 
X=data[features_cols]
y = data.Price

# instantiate, fit
lm =LinearRegression()
lm.fit(X,y)
#
i=0
for col in features_cols:
    print("The Coefficient of ", col , "is :", lm.coef_[i])
    i = i+1
```

# Let's move ahead and check for multicollinearity.

```python
from sklearn.preprocessing import StandardScaler
scaler =StandardScaler()
#X = data[["SqFt","Bedrooms","Bathrooms","Offers","IsBrick","Neighborhood_Direction_North","Neighborhood_Direction_South","Neighborhood_Direction_West"]]

X = data[["SqFt","Bedrooms","Bathrooms","Offers","IsBrick"]]
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = X_scaled

# we create a new data frame which will include all the VIFs
# note that each variable has its own variance inflation factor as this measure is variable specific (not model specific)
# we do not include categorical values for mulitcollinearity as they do not provide much information as numerical ones do
vif = pd.DataFrame()
# here we make use of the variance_inflation_factor, which will basically output the respective VIFs 
vif["VIF"] = [variance_inflation_factor(variables, i) for i in range(variables.shape[1])]
# Finally, I like to include names so it is easier to explore the result
vif["Features"] = X.columns
```

```python
variables.shape[1]
```

```python
vif
```

###### One recommendation is that if VIF is greater than 5, then the explanatoryvariable given by exog_idx is highly collinear with the other explanatory variables, and the parameter estimates will have large standard errors because of this.
#### Here, we have the correlation values for all the features. As a thumb rule, a VIF value greater than 5 means a very severe multicollinearity. We don't any VIF greater than 5 , so we are good to go.
Great. Let's go ahead and use linear regression and see how good it fits our data.
But first. let's split our data in train and test.

```python
from sklearn.model_selection import train_test_split

# x for independen variable and y for dependent variable 
x_train,x_test,y_train,y_test = train_test_split(X_scaled, y, test_size=0.25,random_state=355)
```

```python
print(y_train)
print(y_test, "size is :", y_test.size)
```

```python
regression = LinearRegression()

regression.fit(x_train,y_train)
```

```python
regression.score(x_train,y_train)
```

```python
# Let's create a function to create adjusted R-Squared
def adj_r2(x,y):
    r2 = regression.score(x,y)
    n = x.shape[0]
    p = x.shape[1]
    adjusted_r2 = 1-(1-r2)*(n-1)/(n-p-1)
    return adjusted_r2
```

```python
adj_r2(x_train,y_train)
```

Our r2 score is 66.5% and adj r2 is 63.5% for our training et., so looks like we are not being penalized by use of any feature.

Let's check how well model fits the test data.

Now let's check if our model is overfitting our data using regularization.

```python
regression.score(x_test,y_test)
```

```python
adj_r2(x_test,y_test)
```

So it looks like our model r2 score is more on the test data.
