# Create col6 which keeps data of log10 of values greater than 50 of col2 and if value of col2 is less than 50 then divide it by 10 and fill it in col6

```python
def test(data):
    if data>50:
        return np.log10(data)
    else:
        return data/10
```

```python
test(44)
```

```python
test(55)
```

```python
#but now see the magic
df5['col6']=df5['col2'].apply(test)
```
